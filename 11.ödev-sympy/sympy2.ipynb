```python
# -*- coding: utf-8 -*-
"""
Created on Thu Apr  9 19:59:40 2020

@author: Ozgur Kucet
"""
from sympy import Symbol,pprint
import sympy as sym
import sympy.plotting as syp
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook

sigma = Symbol('sigma')
mu = Symbol('mu')
x = Symbol('x')

#pprint(2*sym.pi*sigma)
#print(sym.sqrt(2*sym.pi*sigma))
#pprint(sym.sqrt(2*sym.pi*sigma))

part_1 = 1/(sym.sqrt(2*sym.pi*sigma**2))
part_2 = sym.exp(-1*((x-mu)**2)/(2*sigma**2))
%matplotlib notebook
my_gaussfunction = part_1*part_2


#pprint(part_1)
#pprint(part_2)
#pprint(my_gaussfunction)

gauss_function = 1/(sym.sqrt(2*sym.pi*sigma))
gauss_function.subs({mu:0,sigma:1})
#print(gauss_function)

syp.plot(my_gaussfunction.subs({mu:1,sigma:3}),(x,-10,10),title="gauss distribution")
```

```python
x_values = []
y_values = []

for value in range(-5,5):
    y=my_gaussfunction.subs({mu:10,sigma:30,x:value})
    y_values.append(y)
    x_values.append(value)
    #print(value,y)

plt.plot(x_values,y_values)
```
